```python
rating=pd.read_csv("ratings_small.csv")
```

```python
rating
```

```python
rating_movies=rating.groupby("movieId", as_index=False)["rating"].median()
```

```python
sns.histplot(data=rating_movies, x="rating")
```

```python
rating_movies["good_movies"]=rating_movies["rating"]>=3
```

```python
rating_movies["good_movies"].value_counts().plot(kind="pie",autopct='%1.1f%%',startangle=90,figsize=(5,5))
```

Tergambarkan dalam pie chart ada 78,4% film bagus dengan rating diatas 3.0

```python
new_df=profitable_movies.merge(rating_movies,
                        left_on="id",right_on="movieId",
                        how="inner")
```

```python
sns.histplot(data=new_df,
             x="profit",
             hue="good_movies")
```

```python
sns.scatterplot(data=new_df, x="runtime", y="profit", hue="good_movies")
```

```python
sns.scatterplot(data=new_df, x="rating", y="profit")
plt.axvline(x=3, color='red', linestyle='--', label='Batas Rating = 3')
plt.axhline(y=0.5e9, color='green', linestyle='--', label='Batas Profit = 0.5e9')
plt.show()
```

Rating bagus tidak menjamin profit tinggi, tetapi film dengan rating buruk lebih jarang menghasilkan profit besar.
